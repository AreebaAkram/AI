```python
import requests
# fetch data / allows you to make  arequest to web page
from bs4 import BeautifulSoup 
# used for scraping
response = requests.get("https://en.wikipedia.org/wiki/Main_Page")
print(response)

content = response.text
print(content)

blob = BeautifulSoup(content, 'html.parser')
data = blob.find_all('h2')
for i in data:
    print(i.get_text())
```
